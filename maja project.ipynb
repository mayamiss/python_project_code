## <center> Tipovi kliničkih studija

* **Interventional studies *(Clinical trials)***: tip kliničkih studija u kojima su ispitanici podeljeni u grupe koje primaju ispitivanu supstancu i one koje primaju placebo, u cilju određivanja efekata ispitivane supstance. Sve analize koje se vrše tokom studije su definisane protokolom 
* **Observational studies**: tip kliničkih studija u kojima su ispitanici svrstani u grupe na osnovu zdravstvenog stanja, odnosno prisustva/odsustva bolesti. Ispitanici mogu primati određenu terapiju, može im se izvršiti intervencija ili određena analiza, ali to ne predstavlja glavne kriterijume na osnovu kojih se vrši podela u grupe. *PRIMER*: *registar pacijenata*
* **Expanded access *(Compassionate use)***: Namenjen pacijentima koji usled bolesti ne mogu biti učesnici u kliničkim studijama. Na ovaj način omogućen im pristup lekovima koji nisu odobreni od strane FDA *(Food and Drug Administration)*  

```python
observational = data['Study Type'].value_counts()['Observational']
interventional = data['Study Type'].value_counts()['Interventional']
expanded_access = data[data['Study Type'].str.contains('Expanded Access',regex=False)]
slices = [observational,interventional, expanded_access['Study Type'].count()]
plt.title("Tipovi studija", fontdict = {'fontsize':20})
labels=["Observational","Interventional", "Expanded Access"]
colors = ["r","b", "g"]

plt.pie(slices,labels = labels,colors=colors,startangle=90,explode=(0.1,0.1,0.1), shadow=True,autopct="%2.2f %%")
plt.legend(bbox_to_anchor=(1, 0, 0.5, 1))
plt.show()
```

## <center> Zastupljenost intervencijskih i obzervacionih studija po zemljama

```python
labels = data['Country'].value_counts().index
temp_data = data[(data['Study Type'] == 'Interventional') | (data['Study Type'] == 'Observational')]['Country'].value_counts().index

recnik = {}

for country in temp_data:
    inter = data[(data['Study Type'] == 'Interventional') & (data['Country'] == country)]['Country'].value_counts().values
    obser = data[(data['Study Type'] == 'Observational') & (data['Country'] == country)]['Country'].value_counts().values
    tot = int(inter[0]) + int(obser[0])
    inter = round((float(inter)/tot) * 100.00,2)
    obser = round((float(obser)/tot) * 100.00,2)
    recnik[country] = (inter, obser)



niz = np.array(list(recnik.values()))
niz_cum = niz.cumsum(axis=1)

fig, ax = plt.subplots(figsize=(10, 10))
#ax.invert_yaxis()
ax.xaxis.set_visible(False)
ax.set_xlim(0, np.sum(niz, axis=1).max())


category_names = ['Interventional', 'Observational']
category_colors = plt.get_cmap('RdYlGn')(
    np.linspace(0.15, 0.85, niz.shape[1]))



for i, (colname, color) in enumerate(zip(category_names, category_colors)):
    widths = niz[:, i]
    print(widths)
    starts = niz_cum[:, i] - widths
    ax.barh(labels, widths, left=starts, height=0.7, label=colname, color=color)
    xcenters = starts + widths / 2

    r, g, b, _ = color
    text_color = 'black' if r * g * b < 0.5 else 'darkgrey'
    for y, (x, c) in enumerate(zip(xcenters, widths)):
        ax.text(x, y, str(float(c)) + " %", ha='center', va='center', color=text_color)
    ax.legend(ncol=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize= 15)


plt.show()
```

***ZAKLJUČAK***

***U odnosu na ukupan broj studija, tri države u kojima su najzastupljenije intervencijske studije su SAD, Brazil i Meksiko, dok su obzervacione najzasputljenije u Turskoj, Francuskoj i Egiptu***

## <center> Faze COVID-19 klinickih studija

* **Faza 1:** se obično sprovodi kod malog broja isptanika koji su najčešće ***zdravi dobrovoljci***. U toj fazi ispituje se doza leka, način primene leka, njegova raspoloživost u organizmu, podnošljivost i neželjeni događaji. ***Cilj ove faze je da se ustanovi bezbednost primene leka.***
* **Faza 2:**  je etapa u razvoju leka u kojoj se ***ispituje optimalna doza***, kod ***većeg broja ispitanika (do nekoliko stotina) koji boluju od bolesti*** za koju je lek namenjen. Prikupljaju se i važni podaci o ***podnošljivosti i bezbednosti primene leka kod bolesnika.***
* **Faza 3:** obično uključuje ***veći broj (nekoliko stotina do više hiljada) ispitanika*** u brojnim centrima i zemljama. Ispituje se ***efikasnost i bezbednost primene leka u poređenju sa standardnim lekovima za tu bolest ili placebom.*** Ako dobijeni podaci pokažu pozitivan rezultat ispitivanja odnosno zadovoljavajući efikasnost pri lečenju i sigurnost pri primeni, podnosi se zahtev za registraciju leka kod regulatornih tela različitih zemalja (u Srbiji je to Agencija za lekove i medicinska sredstva Srbije), posle čega on postaje dostupan za lečenje bolesnika širom sveta.
* **Faza 4:** se sprovodi nakon što je ***lek odobren za primenu***, sa ciljem prikupljanja što većeg broja podataka o bezbednosti i efikasnost leka. U ovoj postregistracionoj fazi se pre svega sprovodi ***praćenje neželjenih efekata*** u svakodnevnoj terapijskoj praksi kod široke populacija.

```python
ep1 = data['Phases fill'].value_counts()['Early Phase 1']
not_app = data['Phases fill'].value_counts()['Not Applicable']
p1 = data['Phases fill'].value_counts()['Phase 1']
p1p2 = data['Phases fill'].value_counts()['Phase 1|Phase 2']
p2 = data['Phases fill'].value_counts()['Phase 2']
p2p3 = data['Phases fill'].value_counts()['Phase 2|Phase 3']
p3 = data['Phases fill'].value_counts()['Phase 3']
p4 = data['Phases fill'].value_counts()['Phase 4']
labels = ['Early Phase 1','Not Applicable', 'Phase 1', 'Phase 1|Phase 2', 'Phase 2', 'Phase 2|Phase 3', 'Phase 3', 'Phase 4']
slices = [ep1,not_app, p1, p1p2, p2, p2p3, p3, p4]
colors = ["r","b", "g", "#FFFF00","#A0A0A0", "#CC00CC", "#66FF66", "#FF007F"  ]
explode = (0, 0, 0, 0, 0, 0, 0, 0)  # explode a slice if required

plt.pie(slices, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True)
plt.legend(bbox_to_anchor=(1, 0, 0.5, 1))
        
#draw a circle at the center of pie to make it look like a donut
centre_circle = plt.Circle((0,0),0.4,color='black', fc='white',linewidth=1)
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


# Set aspect ratio to be equal so that pie is drawn as a circle.
plt.axis('equal')
plt.show()  
```

```python
labels = ['Not Applicable','Early Phase 1', 'Phase 1', 'Phase 1|Phase 2', 'Phase 2', 'Phase 2|Phase 3', 'Phase 3', 'Phase 4']

inter_nap = sum(data[(data['Phases fill']=='Not Applicable') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_nap = sum(data[(data['Phases fill']=='Not Applicable') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
          
inter_ep1 = sum(data[(data['Phases fill']=='Early Phase 1') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_ep1 = sum(data[(data['Phases fill']=='Early Phase 1') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)

inter_p1 = sum(data[(data['Phases fill']=='Phase 1') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p1 = sum(data[(data['Phases fill']=='Phase 1') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
          
inter_p1p2 = sum(data[(data['Phases fill']=='Phase 1|Phase 2') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p1p2 = sum(data[(data['Phases fill']=='Phase 1|Phase 2') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
          
inter_p2 = sum(data[(data['Phases fill']=='Phase 2') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p2 = sum(data[(data['Phases fill']=='Phase 2') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
          
inter_p2p3 = sum(data[(data['Phases fill']=='Phase 2|Phase 3') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p2p3 = sum(data[(data['Phases fill']=='Phase 2|Phase 3') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
          
inter_p3 = sum(data[(data['Phases fill']=='Phase 3') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p3 = sum(data[(data['Phases fill']=='Phase 3') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)
        
inter_p4 = sum(data[(data['Phases fill']=='Phase 4') & (data['Study Type']=='Interventional')]['Phases fill'].value_counts().values)
observ_p4 = sum(data[(data['Phases fill']=='Phase 4') & (data['Study Type']=='Observational')]['Phases fill'].value_counts().values)

inter=[inter_nap, inter_ep1, inter_p1, inter_p1p2, inter_p2, inter_p2p3, inter_p3, inter_p4]
obser=[observ_nap, observ_ep1, observ_p1, observ_p1p2, observ_p2, observ_p2p3, observ_p3, observ_p4]


x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots(figsize=(12,10))
rects1 = ax.bar(x - width/2, inter, width, label='Interventional')
rects2 = ax.bar(x + width/2, obser, width, label='Observational')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Broj studija' ,fontdict = {'fontsize':15})
ax.set_title('Faze studija u odnosu na tip studija',fontdict = {'fontsize':20})
ax.set_xticks(x)
ax.set_xticklabels(labels, fontdict = {'fontsize':10})
ax.legend()
ax.grid(True)

def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

***ZAKLJUČAK***

* ***Zastupljenost intervencijskih u odnosu na obzervacione studije je veca kod studija druge i trece faze, sto je i ocekivano, jer su u tim fazama ispitanici podeljeni u grupe (ispitivana supstanca vs. placebo), po definisanom protokolu, u cilju određivanja efekata ispitivane supstance.***
* ***Nasuprot tome, u studijama cetvrte faze, zastupljenije su obzervacione studije, jer se tada vrsi pracenje (observacija) potencijalnih nezeljenih efekata leka na velikoj populaciji, kada je lek vec registrovan i pusten na trziste.***
